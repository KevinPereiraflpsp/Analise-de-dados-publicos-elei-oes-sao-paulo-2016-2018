```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
abst16= pd.read_csv(r"C:\Users\admin\Desktop\Trabalho\2016\Abstenção\perfil_comparecimento_abstencao_2016.csv", encoding= 'latin-1', sep= ';')
```

```python
abst16.info
```

```python
selabst16= ['NR_TURNO', 'SG_UF', 'NM_MUNICIPIO', 'DS_GENERO', 'DS_ESTADO_CIVIL', 'DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE', 'QT_APTOS', 'QT_COMPARECIMENTO', 'QT_ABSTENCAO', 'QT_COMPARECIMENTO_DEFICIENCIA', 'QT_ABSTENCAO_DEFICIENCIA', 'QT_COMPARECIMENTO_TTE', 'QT_ABSTENCAO_TTE']
```

```python
abst16sel= abst16.filter(items=selabst16)
```

```python
abst16sel.head()
```

```python
abst16sel= abst16sel[abst16sel['SG_UF'].str.contains('SP', na= False)]
```

```python
turno1= abst16sel.query('NR_TURNO== "1"')
turno2= abst16sel.query('NR_TURNO== "2"')
```

```python
genero1= abst16sel.query('NR_TURNO== "1"')
genero2= abst16sel.query('NR_TURNO== "2"')
genero1= genero1['DS_GENERO'].value_counts(normalize=True)
genero2= genero2['DS_GENERO'].value_counts(normalize=True)
tabgenero1= pd.DataFrame([genero1], index=['Turno 1']).T
tabgenero1.plot(kind= 'bar', title= 'Gênero dos eleitores no 1º Turno - 2016')
tabgenero1['Turno 1'].value_counts(normalize=True)*100
tabgenero1['Turno 1']*100
```

```python
tabgenero2= pd.DataFrame([genero2], index=['Turno 2']).T
tabgenero2.plot(kind= 'bar', title= 'Gênero dos eleitores no 2º turno - 2016', color= ['red'])
tabgenero2['Turno 2']*100
```

```python
civil1= abst16sel.query('NR_TURNO== "1"')
civil2= abst16sel.query('NR_TURNO== "2"')
civil1= civil1['DS_ESTADO_CIVIL'].value_counts(normalize=True)
civil2= civil2['DS_ESTADO_CIVIL'].value_counts(normalize=True)
tabcivil1= pd.DataFrame([civil1], index=['Turno 1']).T
tabcivil1.plot(kind= 'bar', title= 'Estado civil dos eleitores no 1º turno - 2016')
tabcivil1['Turno 1'].value_counts(normalize=True)*100
tabcivil1['Turno 1']*100
```

```python
tabcivil2= pd.DataFrame([civil2], index=['Turno 2']).T
tabcivil2['Turno 2'].value_counts(normalize=True)*100
tabcivil2.plot(kind= 'bar', title= 'Estado civil dos eleitores no 2º turno - 2016', color= ['red'])
tabcivil2['Turno 2']*100
```

```python
idade1= abst16sel.query('NR_TURNO== "1"')
idade2= abst16sel.query('NR_TURNO== "2"')
idade1= idade1['DS_FAIXA_ETARIA'].value_counts(normalize=True)
idade2= idade2['DS_FAIXA_ETARIA'].value_counts(normalize=True)
tabidade1= pd.DataFrame([idade1], index=['Turno 1']).T
tabidade1.plot(kind= 'bar', title= 'Faixa Etária dos eleitores no 1º turno - 2016')
tabidade1['Turno 1'].value_counts(normalize=True)*100
tabidade1['Turno 1']*100
```

```python
tabidade2= pd.DataFrame([idade2], index=['Turno 2']).T
tabidade2.plot(kind= 'bar', title= 'Faixa Etária dos eleitores no 2º turno - 2016', color= ['red'])
tabidade1['Turno 1'].value_counts(normalize=True)*100
tabidade1['Turno 1']*100
```

```python
escolaridade1= abst16sel.query('NR_TURNO== "1"')
escolaridade2= abst16sel.query('NR_TURNO== "2"')
escolaridade1= escolaridade1['DS_GRAU_ESCOLARIDADE'].value_counts(normalize=True)
escolaridade2= escolaridade2['DS_GRAU_ESCOLARIDADE'].value_counts(normalize=True)
tabescolaridade1= pd.DataFrame([escolaridade1], index=['Turno 1']).T
tabescolaridade1.plot(kind= 'bar', title= 'Grau de escolaridade dos eleitores no 1º turno - 2016')
tabescolaridade1['Turno 1'].value_counts(normalize=True)*100
tabescolaridade1['Turno 1']*100
```
